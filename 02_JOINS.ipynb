### JOIN operations


Inner Join

```python
SELECT 
    a.appointment_id,
    a.appointment_date,
    a.status,
    p.first_name,
    p.last_name
FROM appointments a
INNER JOIN patients p ON a.patient_id = p.patient_id;
```

Left Join

```python
SELECT 
    p.patient_id,
    p.first_name,
    p.last_name,
    a.appointment_id,
    a.appointment_date,
    a.status
FROM patients p
LEFT JOIN appointments a ON p.patient_id = a.patient_id;
```

Right Join


```python
SELECT p.first_name, a.appointment_id, a.status
FROM patients p
RIGHT JOIN appointments a ON p.patient_id = a.patient_id;
```

Cross Join


```python
SELECT p.first_name, a.appointment_id
FROM patients p
CROSS JOIN appointments a;
```

Full Outer Join

```python
SELECT p.first_name, a.appointment_id, a.status
FROM patients p
FULL OUTER JOIN appointments a ON p.patient_id = a.patient_id;
```

Self join

```python
SELECT p1.patient_id AS patient1, p2.patient_id AS patient2, a1.doctor_id
FROM appointments a1
JOIN appointments a2 ON a1.doctor_id = a2.doctor_id
JOIN patients p1 ON a1.patient_id = p1.patient_id
JOIN patients p2 ON a2.patient_id = p2.patient_id
WHERE p1.patient_id <> p2.patient_id;
```
